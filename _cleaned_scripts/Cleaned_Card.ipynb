```python
import pandas as pd

# Read the card dataset
card_df = pd.read_csv('card.csv')

# Display basic information
print('Dataset shape:', card_df.shape)
print('\nFirst few rows:')
card_df.head()
```

```python
# Replace NaN/empty values in country, province, and city columns with "other"
print("Before cleaning:")
print(f"Country - Total count: {len(card_df)}, NaN count: {card_df['country'].isna().sum()}")
print(f"Province - Total count: {len(card_df)}, NaN count: {card_df['province'].isna().sum()}")
print(f"City - Total count: {len(card_df)}, NaN count: {card_df['city'].isna().sum()}")

# Fill NaN values with "other"
card_df['country'] = card_df['country'].fillna('other')
card_df['province'] = card_df['province'].fillna('other')
card_df['city'] = card_df['city'].fillna('other')

print("\nAfter cleaning:")
print(f"Country - Total count: {len(card_df)}, NaN count: {card_df['country'].isna().sum()}")
print(f"Province - Total count: {len(card_df)}, NaN count: {card_df['province'].isna().sum()}")
print(f"City - Total count: {len(card_df)}, NaN count: {card_df['city'].isna().sum()}")

print("\nSample of cleaned data:")
card_df[['country', 'province', 'city']].head(10)
```

```python
# Save the cleaned dataframe to Card_Cleaned.csv
card_df.to_csv('Card_Cleaned.csv', index=False)
print(f"Cleaned dataset saved to 'Card_Cleaned.csv'")
print(f"Shape: {card_df.shape}")
```
