```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("E:/After 1-16 Cleaned/Cleaned_Original/Cleaned_cheque.csv")

print(f"Dataset loaded: {len(df)} rows")
print(f"Columns: {df.columns.tolist()}")
print(f"\nFirst few rows:")
df.head()
```

```python
# Check the amount_cad column before multiplication
print("Amount CAD column info:")
print(f"  Data type: {df['amount_cad'].dtype}")
print(f"  Sample values before multiplication:")
print(df['amount_cad'].head(10))
print(f"\n  Statistics:")
print(df['amount_cad'].describe())
```

```python
# Multiply all cells in the amount_cad column by 100 and round to whole numbers (cents)
# Rounding eliminates floating-point precision errors
df['amount_cad'] = (df['amount_cad'].round(0))

print("Amount CAD column multiplied by 100 and rounded")
print(f"\nSample values after multiplication:")
print(df['amount_cad'].head(10))
print(f"\nStatistics after multiplication:")
print(df['amount_cad'].describe())
```

```python
# Save the modified dataset back to the same CSV file
df.to_csv("E:/After 1-16 Cleaned/Cleaned_Original/Cleaned_cheque.csv", index=False)
print(f"Total rows saved: {len(df)}")
```
