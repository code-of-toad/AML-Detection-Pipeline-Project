```python
import pandas as pd
```

```python
# Read the CSV files
individual_df = pd.read_csv('individual_cleaned.csv')
business_df = pd.read_csv('business_cleaned.csv')

print(f"Individual records: {len(individual_df)}")
print(f"Business records: {len(business_df)}")
```

```python
# Replace NaN and "NA" with "other" in province column
individual_df['province'] = individual_df['province'].fillna('other').replace('NA', 'other')
business_df['province'] = business_df['province'].fillna('other').replace('NA', 'other')

print("NA values replaced with 'other'")
print(f"Individual - 'other' count: {(individual_df['province'] == 'other').sum()}")
print(f"Business - 'other' count: {(business_df['province'] == 'other').sum()}")
```

```python
# Save the modified files
individual_df.to_csv('individual_cleaned_Jan16.csv', index=False)
business_df.to_csv('business_cleaned_Jan16.csv', index=False)

print("Files saved successfully!")
print("  - individual_cleaned_Jan16.csv")
print("  - business_cleaned_Jan16.csv")
```

```python
# Verify the distribution after replacement
print("=" * 60)
print("PROVINCE DISTRIBUTION - INDIVIDUAL CLEANED (After replacement)")
print("=" * 60)
print(individual_df['province'].value_counts())
print(f"\nTotal: {len(individual_df)}")
```

```python
# Verify the distribution after replacement
print("=" * 60)
print("PROVINCE DISTRIBUTION - BUSINESS CLEANED (After replacement)")
print("=" * 60)
print(business_df['province'].value_counts())
print(f"\nTotal: {len(business_df)}")
```
